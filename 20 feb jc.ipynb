# Arithmetic operator

```python
a = 5
b = 2
sum_result = a + b
difference_result = a - b
product_result = a * b
division_result = a / b
modulus_result = a % b
exponentiation_result = a ** b
print('sum : ',sum_result)
print('difference : ',difference_result)
print('product : ',product_result)
print('division : ',division_result)
print('modulus :',modulus_result)
print('exponential :',exponentiation_result)
```

# comparison operator

```python
is_equal = a == b
is_not_equal = a != b
is_greater_than = a > b
is_less_than = a < b
print('is equal : ',is_equal)
print('is not equal : ',is_not_equal)
print('is greater than : ',is_greater_than)
print('is less than: ',is_less_than)
```

# logical operator

```python
logical_and = (a > 0) and (b > 0)
logical_or = (a > 0) or (b > 0)
logical_not = not (a > 0)
print('logical and : ',logical_and)
print('logical or : ',logical_or)
print('logical not : ',logical_not)
```

# unary operator

```python
number = 10
result = -number
print('unary minus operator :',result)
result = +number
print('unary plus operator :',result)
is_true = True
logical_not_result = not is_true
print('logical not operator :',logical_not_result)
```

# Binary operator

```python
a = 5
b = 3
addition_result = a + b
print('Binary addition result : ', addition_result)
multiplication_result = a * b
print('Binary multiplication result : ', multiplication_result)
bitwise_and_result = a & b
print('Bitwise and result : ', bitwise_and_result)
```
