# superset | subset | disjoint

```python
s5 = {1,2,3,4,5,6,7,8,9}
s6 = {3,4,5,6,7,8}
s7 = {10,20,30,40}
```

```python
s5.issuperset(s6)
```

```python
s6.issuperset(s5)
```

```python
s6.issubset(s5)
```

```python
s7.isdisjoint(s5)
```

```python
s5.isdisjoint(s6)
```
