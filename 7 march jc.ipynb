# Array creation function

```python
import numpy as np
```

```python
# create an array from list
a = np.array([1,2,3])
print("Array a: ",a)
```

```python
# create an array with evenly spaced values
b = np.arange(0,10,2) # values from 0 to 10 with step 2
print("Array b: ",b)
```

```python
# create an array filled with zeros
d = np.zeros((2,3))
print("Array d:\n",d)
```

```python
d = np.zeros((2,3),dtype=int)
print("Array d:\n",d)
```

```python
# create an array filled with ones
e = np.ones((3,2),dtype=int)
print("Array e:\n",e)
```

```python
# create an identity matrix
f = np.eye(4) #4*4 identity matrix
print("Identity matrix f:\n",f)
```

# array manipulation function

```python
# reshape an array
a1 = np.array([1,2,3])
reshaped = np.reshape(a1,(3,1))
print(reshaped)
```

```python
# flatten an array
f1 = np.array([[1,2,3],[3,4,5],[5,6,7]])
flattend = np.ravel(f1)
print(flattend)
```

```python
# transpose an array
e1 = np.array([[1,2],[3,4]])
print(e1)
transposed = np.transpose(e1)
print("transposed :\n",transposed)
```

```python
# stack arrays vertically
a2 = np.array([1,2])
b2 = np.array([3,4])
stacked = np.vstack([a2,b2])
print(stacked)
```

```python
a2 = np.array([1,2])
b2 = np.array([3,4])
stacked = np.hstack([a2,b2])
print(stacked)
```

# mathematical function

```python
# add two array
g = np.array([1,2,3,4])
added = np.add(g,2)
print(added)
```

```python
# square each element
squared = np.power(g,2)
print(squared)
```

```python
# square root of each element
sqrt_val = np.sqrt(g)
print(sqrt_val)
```

# statistical function

```python
s = np.array([1,2,3,4])
mean = np.mean(s)
print(mean)
```

```python
s = np.array([1,2,3,4])
mean = np.var(s)
print(mean)
```

```python
std_dev = np.std(s)
print(std_dev)
```

```python
minimum = np.min(s)
print(minimum)
```

```python
maximum = np.max(s)
print(maximum)
```

# random sampling function

```python
# generate random values between 0 and 1
random_vals = np.random.rand(3)
print(random_vals)
```

```python
# generate random integers
rand_ints = np.random.randint(0,10,size=5)
print(rand_ints)
```

# boolean and logical


```python
# check if all elements are true
logical_test = np.array([True,False,True])
all_true = np.all(logical_test)
print(all_true)
```

```python
logical_test = np.array([False,False,False])
all_true = np.all(logical_test)
print(all_true)
```

```python
logical_test = np.array([True,False,False])
any_true = np.any(logical_test)
print(any_true)
```

# set operations

```python
# intersection of two arrays
a = np.array([1,2,3,4])
b = np.array([3,4,5,6])
intersection = np.intersect1d(a,b)
print(intersection)
```

```python
union = np.union1d(a,b)
print(union)
```

# array attribute functions

```python
a = np.array([1,2,3])
shape = a.shape
size = a.size
dimensions = a.ndim
dtype = a.dtype
print(shape)
print(size)
print(dimensions)
print(dtype)
```

# other functions

```python
# create a copy of array
a1 = np.array([1,2,3,4])
copied_array = np.copy(a1)
print(copied_array)
```

```python
# size in bytes of an array
a1 = a.nbytes
print(a1)
```

```python
# check if two array share memory
shared = np.shares_memory(a1,copied_array)
print(shared)
```
