```python
num1 = 100
num2 = 25
num3 = 333
avg = (num1 + num2 +num3) / 3
print('The avg of {},{},and{} is = {}' .format(num1,num2,num3,avg))
```

```python
num1 = 100
num2 = 25
num3 = 333
avg1 = round((num1 + num2 + num3) / 3,2) #round off till 2 digits
print('The avg of {},{},and{} is = {}' .format(num1,num2,num3,avg1))
```
