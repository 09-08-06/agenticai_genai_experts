# LOOPS

```python
i = 1
while i <= 5:
    print('data science')
    i = i + 1
```

```python
i = 1
while i <=5:
    print('data science')
    j = 1
    while j <=4:
        print('technology')
        j = j + 1
    i = i + 1
    print()
```

```python
i = 1
while i <= 5:
    print('data science',end= " ")
    j = 1
    while j <= 4:
        print('technology',end= " ")
        j = j + 1
    i = i + 1
    print()
```
